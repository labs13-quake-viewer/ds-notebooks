```python
damage = pd.read_csv('https://raw.githubusercontent.com/labs13-quake-viewer/ds-data/danielle-romanoff/earthquake_damage.csv')
damage.head()
```

```python
damage.shape
```

```python
damage.Magnitude.isnull().sum()
```

```python
damage = damage.drop(0)
damage.head()
```

```python
damage.Mo = damage.Mo.fillna(1.0)
damage.Dy = damage.Dy.fillna(1.0)
damage.head()
```

```python
damage['Date'] = damage[damage.columns[0:3]].apply(
    lambda x: ','.join(x.dropna().astype(int).astype(str)),
    axis=1
)
damage.head()
```

```python
damage['Date'] = damage['Date'].astype(str)
damage.head()
```

```python
damage['Date'] = pd.to_datetime(damage['Date'], infer_datetime_format=True)
damage.head()                                        
```
